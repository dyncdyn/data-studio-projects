## Plastic waste exported by the US (including district-level) from 2014 to 2019, yearly and monthly

```python
df = pd.read_csv('StandardReportExports.csv', skiprows = 4, index_col = False, thousands=',')
df = df.drop(columns = ['Unnamed: 4'])
df['Value (million$US)'] = round(df['Value ($US)']/1000000, 2)
df['Value (thousand$US)'] = round(df['Value ($US)']/1000, 2)
df
```

```python
df_all = df[df['District'] == 'All Districts'].reset_index().drop(columns = 'index')

alist = ['2014','2015','2016','2017','2018']
df_year = df_all[df_all['Time'].isin(alist)].reset_index().drop(columns = 'index')
df_year['Value ($US)'] = round(df_year['Value ($US)']/1000000, 2)
df_year
```

### Total Value of Plastic Waste in US Exports, Yearly (2014-2018)

```python
ax = df_year.plot(x='Time', y='Value ($US)',
                  grid = True, 
                  style='*-', 
                  figsize=(10, 5))

plt.title("Total Value of Plastic Waste in US Exports, Yearly(2014-2018)", pad = 20)
plt.tight_layout()
plt.savefig('output1.pdf')
```

### Total Value of Plastic Waste in US Exports, Monthly (2014-2019.5)

```python
alist = ['2014','2015','2016','2017','2018','2019 through May']
df_month = df_all[~df_all['Time'].isin(alist)]
df_month
```

```python
df_month.plot(x = 'Time', y = 'Value (million$US)', figsize = (30,10), grid = True, style = '*-')
plt.title("Total Value of Plastic Waste in US Exports, Monthly (2014-2019.5)",fontsize = 30, pad = 20)

plt.tight_layout()
plt.savefig('output2.pdf')
```

### Exported Plastic Waste of New York City in 2018

```python
df_nyc2_2018 = df_nyc2[df_nyc2['Time'] == '2018']
df_nyc2_2018
```

```python
continents = ['Africa','Asia','Europe','Australia and Oceania','North America','South/Central America']
continentsX = ['All Geographic Regions (World Total)','Africa','Asia','Europe','Australia and Oceania','North America','South/Central America']
subcont = ['Asia Near East','Asia - South','Asia - Other','Central America and Carribean','South America']
df_nyc2_c = df_nyc2_2018[~df_nyc2_2018['Country'].isin(continentsX)]
df_nyc2_c = df_nyc2_c[~df_nyc2_c['Country'].isin(subcont)].reset_index().drop(columns = 'index')

df_nyc2_c
```

```python
ax = df_nyc2_c.sort_values(by = 'Value ($US)', ascending = False).head(10).\
    plot(kind = 'barh', x = 'Country', y ='Value ($US)', figsize = (10,5))
for i in ax.patches:
    ax.text(i.get_width(), i.get_y()+.41, \
            str(i.get_width()), fontsize=10, color='grey')
    
plt.gca().invert_yaxis()

plt.title("2018", fontsize = 12, pad = 20)
plt.tight_layout()
plt.savefig('output6.png')
```
