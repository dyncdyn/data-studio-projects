## Make a map of the number of cases involved wiretaps in district courts of the U.S. in 2018

```python
df2 = pd.read_csv('geo.csv', index_col = 0)
df2
```

```python
DF = df.merge(df2, on = 'Jurisdiction')
DF
```

```python
DF['freq'] = DF.groupby('Jurisdiction')['Jurisdiction'].transform('count')
DF
```

```python
DF_not_rp = DF.drop_duplicates('Jurisdiction')
DF_not_rp.to_csv('wiretap_counts.csv')
```

```python
df.columns
```

```python
DF_not_rp = DF_not_rp.sort_values(['freq'], ascending = False).reset_index(drop=True)
DF_not_rp5 = DF_not_rp.head()
DF_not_rp5
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(data=DF_not_rp5, x = 'freq', y = 'Jurisdiction' , color='lightgrey')

for i in ax.patches:
    ax.text(i.get_width()+.1, i.get_y()+.31, \
            str(round((i.get_width()), 2)), fontsize=10, color='dimgrey')
plt.title('Top 5 District Courts with the Most Wiretap Cases', fontsize = 15)

plt.tight_layout()
plt.savefig('output1.pdf')
```

```python
df['Number of Incriminating Intercepts'] = df['Number of Incriminating Intercepts'].astype(int)
df['Number of Intercepts'] = df['Number of Intercepts'].astype(int)
df['Intercept%'] = round (df['Number of Incriminating Intercepts']/df['Number of Intercepts'] * 100, 2) 
df
```

```python
data1 = df.groupby(by='Jurisdiction')['Intercept%'].mean()
df_intercept = pd.DataFrame(data1)
```

```python
df_intercept.reset_index(level=0, inplace=True)
df_intercept
```

```python
df_intercept = df_intercept.replace([np.inf, -np.inf], np.nan).dropna()
df_intercept = df_intercept[df_intercept['Intercept%'] <= 100]
df_intercept['Intercept%'] = round(df_intercept['Intercept%'],2)
```

```python
plt.figure(figsize = (20,10))
sns.distplot(df_intercept['Intercept%'], bins = 100)

plt.title('Distribution of the Effective Percentages of Incriminating Intercepts', fontsize = 20, pad = 10)

plt.tight_layout()
plt.savefig('output3.pdf')
```

```python
df_intercept1 = df_intercept.sort_values(['Intercept%'], ascending = False).reset_index(drop=True)
df_intercept1
```

```python
# sns.barplot(x = 'Intercept%', y = 'Jurisdiction')

plt.figure(figsize = (20,20))
ax = sns.barplot(data = df_intercept1,x = 'Intercept%', y = 'Jurisdiction', color='lightblue')

for i in ax.patches:
    ax.text(i.get_width()+.1, i.get_y()+.61, \
            str(((i.get_width()))), fontsize=10, color='dimgrey')
plt.title('Effective percentage of Incriminating Intercepts', fontsize = 20, pad =10)

plt.tight_layout()
plt.savefig('output4.pdf')
```
