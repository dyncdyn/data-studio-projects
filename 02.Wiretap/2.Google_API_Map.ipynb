```python
import requests
from dotenv import load_dotenv
import os
import pandas as pd
```

```python
df = pd.read_csv('Courts.csv', index_col = 0)
df
```

```python
api_key = os.getenv('Google_API_KEY')
#https://www.youtube.com/watch?v=5iWhQWVXosU
```

```python
courts = df.Jurisdiction.tolist()
```

```python
geocodes = []
for court in courts:
    url = 'https://maps.googleapis.com/maps/api/geocode/json'
    keys = {'address': court, 'key': api_key}
    r = requests.get(url, params=keys)
    
    result_dic = r.json()
    geocode = result_dic['results'][0]['geometry']['location']
    geocodes.append(geocode)
```

```python
geocodes
```

```python
df['geocode'] = geocodes
df
```

```python
df['lat'] = df.geocode.apply(lambda x: x['lat'])
df['lng'] = df.geocode.apply(lambda x: x['lng'])
```

```python
df.to_csv('geo.csv')
```
