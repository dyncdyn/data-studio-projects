```python
import pandas as pd
import requests
import numpy as np
import seaborn as sns
from bs4 import BeautifulSoup
```

```python
url = 'https://www.newpages.com/independent-bookstores/new-york-bookstores'
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
doc
```

## A list of indie bookstores

```python
store = doc.find_all(class_="itemContainer itemContainerLast")
len(store)
```

## A list of indie bookstores' names

> ### question

```python
cats = doc.find_all(class_="catItemExtraFieldsValue")
cats
for cat in cats:
    name = cat.find('a')
    print(name)
```

```python
cats = doc.find_all(class_="catItemExtraFieldsValue")
```

```python
names = []

for cat in cats:
    name = cat.find('a')
    if name is not None:
        name = name.text
        label = ['Facebook','Twitter','Website','Instagram']
        if name not in label:
            names.append(name)
len(names)
```

## Address List

```python
address = doc.find_all(class_="odd typeTextfield group13 ibsAddress")
address
```

```python
addresses = []
for add in address:
    add = add.find(class_="catItemExtraFieldsValue").text
    addresses.append(add)
len(addresses)
```

## Phone list

```python
Phones = []
phone = doc.find_all(class_="even typeTextfield group13 ibsPhone")
phone
for ph in phone:
    ph = ph.find(class_="catItemExtraFieldsValue").text
    Phones.append(ph)
len(Phones)
```
